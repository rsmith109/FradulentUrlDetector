```python
import whois

domain = whois.whois('example.com')

print(domain)
```

```python
import time

def get_single_date(date):
    if isinstance(date, list):
        date = date[0]
    return date

def get_whois_info(url):
    time.sleep(1)

    counter = 1

    if counter % 1000 == 0:
        print(f'Iteration (thousands): {counter // 1000}')
    
    counter += 1
    
    try:
        domain = whois.whois(url)
    except Exception as e:
        print(f'Error Occured: {e}')
        return pd.Series({
            'creation_date': None,
        'updated_date': None,
        'expiration_date': None,
        'domain_name': None,
        'registrar_url': None
        })
    
    creation_date = get_single_date(domain.creation_date)
    updated_date = get_single_date(domain.updated_date)
    expiration_date = get_single_date(domain.expiration_date)
    
    domain_name = domain.domain_name
    registrar_url = domain.registrar_url

    return pd.Series({
        'creation_date': creation_date,
        'updated_date': updated_date,
        'expiration_date': expiration_date,
        'domain_name': domain_name,
        'registrar_url': registrar_url
    })


```

```python
df10k[['creation_date', 'updated_date', 'expiration_date', 'domain_name', 'registrar_url']] = df10k['url'].apply(get_whois_info)
```
